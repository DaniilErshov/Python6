```python
class Count:
    def __init__(self, n):
        self.stop = n
        self.num = 0
    def __iter__(self):
        return self
    def __next__(self):
        self.num += 1
        if self.num <= self.stop: return self.num**2 
        else:
            raise StopIteration
def kv_gen(n):
    for i in range(1,n+1):
        yield i*i
        
c = Count(9)
kv = kv_gen(9)
kv_sp = [i*i for i in range(1, 10)]
for i in range(9):
    print(next(kv))
for i in c:
    print(i)
```

```python

def find_gen(list_1, list_2, to_find):
    for x in list_1:
        for y in list_2:
            result = x * y
            if result == to_find:
                print("found")
                return 
            else:
                yield x, y
    
list_1 = [2, 5, 7, 10]
list_2 = [3, 8, 4, 9]

find = find_gen(list_1, list_2, 56)
for i in find:
    print(i)
```

```python
def Q(n):
    if n == 2 or n == 1:
        return 1
    return Q(n - Q(n - 1)) + Q(n - Q(n - 2))

def Qhoft(lst, N):
    if lst == [1, 1]:
        for i in range(1, N + 1):
            yield Q(i)
            
lst = [1, 1]
N = 10
q = Qhoft(lst, N)
for i in range(1, N + 1):
    print(next(q))
```

```python
class Uzel():
    def __init__(self, n):
        self.info = n
        self.link = None
        
        
class Svaz_spis(Uzel):
    
    def __iter__(self):
        return self
    
    def __init__(self, n = 0):
        self.top = Uzel(n)
        self.curr = self.top
        self.i = 0
        self.end = 1
        
    def append(self, n):
        while self.curr.link != None:
            self.curr = self.curr.link
        self.curr.link = Uzel(n)
        self.end += 1
        self.curr = self.top
        
    def __next__(self):
        self.i += 1
        if self.i == 1:
            self.curr = self.top
            return self.curr.info
        elif self.i <= self.end:
            self.curr = self.curr.link
            return self.curr.info
        else:
            raise StopIteration
            
    def get(self, k):
        i = 1
        self.curr = self.top
        while self.curr.link != None and i < k:
            i += 1
            self.curr = self.curr.link
        if i == k:
            self.curr = self.top
            return self.curr.info
        else:
            self.curr = self.top
            return None
        
    def print_all(self):
        self.curr = self.top
        while (self.curr.link != None):
            print(self.curr.info)
            self.curr = self.curr.link
        print(self.curr.info)
        self.curr = self.top
        
    def remove(self, num):
        self.i = 0
        self.curr = self.top
        if num < self.end:
            if num == 1:
                tmp = self.curr.link
                self.curr.link = None
                self.top = tmp
                del tmp
                self.end -= 1
            else:
                for i in range(1, num - 1):
                    self.curr = self.curr.link
                tmp = self.curr.link.link
                self.curr.link.info = None
                self.curr.link.link = None
                self.curr.link = tmp
                self.end -= 1
                self.i = 0
        elif num == self.end:
            for i in range(1, num - 1):
                self.curr = self.curr.link    
            self.curr.link.info = None
            self.curr.link = None
            self.end -= 1
            self.i = 0
        else:
            raise IndexError
        
uz = Svaz_spis(1)
uz.append(2)
uz.append(3)
uz.append(4)
uz.append(5)
uz.append(6)
uz.append(7)
uz.remove(7)
uz.append(7)

for i in uz:
    print(i)
```
